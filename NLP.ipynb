```python
#install library
import pandas as pd
import nltk
```

```python
#ambil data
df = pd.read_csv("Data Lelang Kendaraan Dua.csv")
df
```

```python
#Drop kolom tidak digunakan
df = df.drop(["NAMA PEMOHON LELANG/PENJUAL", "STATUS LELANG", "JENIS LELANG", "KATEGORI BARANG","NILAI LIMIT" ], axis=1)
```

```python
df.head()
```

```python
#Merubah nama kolom
df = df.rename(columns={"TEMPAT PELAKSANAAN LELANG": "lokasi",
                        "TANGGAL RISALAH LELANG" : "tahun",
                        "POKOK LELANG": "pokok_lelang",
                        "URAIAN BARANG" : "uraian"})
```

```python
df.head()
```

```python
#mengambil informasi tahun
df["tahun"] = pd.to_datetime(df["tahun"])
df["tahun"] = df["tahun"].dt.year
```

```python
df.head()
```

###Proces text analysis

```python
#mengubah huruf menjadi kecil, menghilangkan spasi awal dan akhir, menghilangkan tanda baca
import string
def clean_text(text):
    text = text.lower()
    text = text.strip()
    text = text.translate(str.maketrans("", "", string.punctuation))
    return text

df["uraian"]=df["uraian"].apply(clean_text)
```

```python
df.head(2)
```

```python
#proses tokenizing
import nltk
nltk.download('punkt')
from nltk.tokenize import wordpunct_tokenize
df["uraian"] = df["uraian"].apply(wordpunct_tokenize)
```

```python
df.head(2)
```

```python
df["uraian"].loc[:]
```

```python
#menambahkan kata pada list stopword
new_stopwords = ["1","sepeda","motor", "sepedamotor", "satu", "buah", "kendaraan", "roda", "merk","type", "tipe",
                 "merktipe", "merktype", "unit", "isi silinder", "kondisi", "rusak", "berat", "tahun", "nomor",
                 "mesin", "dalam", "atas", "pembuatan", "isi", "jenis", "perakitan", "model", "silinder",
                 "rangka", "polisi", "warna", "hitam", "bahan", "bakar", "cc", "bensin", "dh", "bermotor",
                 "biru", "merah", "dikuasai"]
nltk.download('stopwords')
stopwords = nltk.corpus.stopwords.words("indonesian") + new_stopwords
stopwords = list(set(stopwords))
```

```python
#proses filtering untuk menghilangkan kata-kata yang tidak meberikan informasi
df["uraian"] = df["uraian"].apply(lambda x: [word for word in x if word not in stopwords])
```

```python
# cek frekueinsi kata dan mengurutkan
from collections import Counter
all_words = [word for sublist in df["uraian"] for word in sublist]
word_counts = Counter(all_words)

```

```python
#mengurutkan hasil dari jumlah tertinggi
word_counts = word_counts.most_common()
word_counts
```

```python
#ambil tipe kendaraan dala kolom uraian
merk_kendaraan = ["honda", "yamaha", "suzuki"]

def cari_merk(row):
    for merk in merk_kendaraan:
        if merk in row:
            return merk
    return None

df['merk_kendaraan'] = df['uraian'].apply(cari_merk)
```

```python
#ambil tipe kendaraan dala kolom uraian2
def cari_merk2(teks):
    return teks[0] if teks else None

df['merk_kendaraan2'] = df['uraian'].apply(cari_merk2)
```
